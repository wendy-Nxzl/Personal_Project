```python
def get_df_from_csv(filename):
  df = pd.read_csv(filename)
  return df

def get_stats(stock):
  df = get_df_from_csv('{}.csv'.format(stock))
  mean = df['Close'].mean()
  std = df['Close'].std()
  cov = std / mean
  return round(mean, 5), round(std, 5), round(cov, 5)
  
stock = "New_AAPL"
stats = get_stats(stock)
print("Stock: {} \n\n- Mean\t\t\t\t: {} \n- Standard deviation\t\t: {} \n- Coefficient of variation\t: {}".format(stock, stats[0], stats[1], stats [2]))
```

```python
import matplotlib.pyplot as plt

data = df['Close']
mean = df['Close'].mean()
std = df['Close'].std()
min_value = min(data)
max_value = max(data)

plt.title("AAPL Dataset")
plt.ylim(min_value - 100, max_value + 100)
plt.scatter(x=df.index, y=df['Close'])
plt.hlines(y=mean, xmin = 0, xmax = len(data))
plt.hlines(y=mean - std, xmin = 0, xmax = len(data), colors = 'r')
plt.hlines(y=mean + std, xmin = 0, xmax = len(data), colors = 'r')
plt.hlines(y=mean - 2*std, xmin = 0, xmax = len(data), colors = 'g')
plt.hlines(y=mean + 2*std, xmin = 0, xmax = len(data), colors = 'g')
plt.show()
```
